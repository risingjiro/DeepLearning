```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x): 
    return 1 / (1 + np.exp(-x))

def tanh(x): 
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def ReLU(x): 
    return np.maximum(0, x)

x = np.arange(-6.0, 6.0, 0.1)
sig = sigmoid(x)
tan_h = tanh(x)
# relu = ReLU(x)


fig, ax = plt.subplots()

# ax.grid()

ax.plot(x, sig, color="blue", label="sigmoid")
ax.plot(x, tan_h, color="orange", label="tanh")
# ax.plot(x, relu, color="red")

ax.set_yticks(np.arange(-1.0, 1.1, 0.5))

ax.legend()
# fig.tight_layout()
plt.show()
```

```python
relu = ReLU(x)
plt.plot(x, relu, label="ReLU")
plt.legend()

# plt.title("ReLU")
# plt.yticks(np.arange(-1.0, 5.0, 0.5))
plt.show()
```
